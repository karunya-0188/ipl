```python
#HANDLE NULL VALUES
matches['city']=matches['city'].fillna("Unknown")
matches['player_of_match']=matches['player_of_match'].fillna("No Award")
matches['winner']=matches['winner'].fillna("No Result")
matches['result_margin']=matches['result_margin'].fillna(0)
matches['method']=matches['method'].fillna("Normal")

matches.isnull().sum()
```

```python
#CHECK NULL VALUES
deliveries.isnull().sum()
```

```python
#HANDLE NULL VALUES
deliveries['extras_type']=deliveries['extras_type'].fillna("Unknown")
deliveries['player_dismissed']=deliveries['player_dismissed'].fillna("Unknown")
deliveries['dismissal_kind']=deliveries['dismissal_kind'].fillna("Unknown")
deliveries['fielder']=deliveries['fielder'].fillna("Unknown")

deliveries.isnull().sum()
```

```python
# Converting date column into proper datetime format
matches['date']=pd.to_datetime(matches['date'],dayfirst=True)
```

```python
# Displaying first few rows of date column after conversion
matches['date'].head()
```

```python
# Creating a new column for match year
matches['year']=matches['date'].dt.year
```

```python
# Displaying date and extracted year columns
matches[['date','year']].head()
```

```python
# Creating match phase categories based on overs
def over_phase(over):
    if over<=5:
        return 'Powerplay'
    elif over<=15:
        return 'Middle Overs'
    else:
        return 'Death Overs'
deliveries['phase']=deliveries['over'].apply(over_phase)   
```

```python
# Displaying over and phase columns
deliveries[['over','phase']].head()
```

```python
# Creating boundary indicator column (4s and 6s)
deliveries['is_boundary']=deliveries['batsman_runs'].apply(
    lambda x:1 if x in [4,6] else 0
) 
```

```python
# Displaying batsman runs and boundary indicator
deliveries[['batsman_runs','is_boundary']].head(15)
```

```python
# Creating four indicator column
deliveries['is_four']=deliveries['batsman_runs'].apply(
    lambda x:1 if x==4 else 0
)
# Displaying batsman runs and four indicator
deliveries[['batsman_runs','is_four']].head(20)
```

```python
# Creating six indicator column
deliveries['is_six']=deliveries['batsman_runs'].apply(
    lambda x:1 if x==6 else 0
)
# Displaying batsman runs and six indicator
deliveries[['batsman_runs','is_six']].head(20)
```

```python
# Creating dot ball indicator column
deliveries['is_dot']=deliveries['batsman_runs'].apply(
    lambda x:1 if x in [0] else 0
) 
```

```python
# Displaying total runs and dot ball indicator
deliveries[['batsman_runs','is_dot']].head(10)
```

```python
# Displaying unique dismissal types line by line
for dismissal in deliveries['dismissal_kind'].unique():
    print(dismissal)
```

```python
# Creating bowler wicket indicator column
deliveries['bowler_wicket']=deliveries['dismissal_kind'].apply(
    lambda x:1
    if x in[
        'bowled',
        'caught',
        'caught and bowled',
        'lbw',
        'stumped',
        'hit wicket'
    ]
    else 0
)
```

```python
# Displaying dismissal kind and bowler wicket indicator
deliveries[['dismissal_kind','bowler_wicket']].head(100)
```

```python
# Counting bowler wickets
deliveries['bowler_wicket'].value_counts()
```
